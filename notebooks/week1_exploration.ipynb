````python
import pandas as pd
import numpy as np

# Load original - this will never be touched
df = pd.read_csv('C:/SupplyChainAssistant/data/supply_chain_data.csv')

# Working copy - all exploration and cleaning happens on this
df_clean = df.copy()

print("Original shape:", df.shape)
print("Copy shape:", df_clean.shape)
print("Both should be identical ✓")
````

````python
missing = df_clean.isnull().sum()
missing_pct = (df_clean.isnull().sum() / len(df)) * 100
missing_report = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage': missing_pct
})
print(missing_report)
````

````python
print("Duplicate rows:", df_clean.duplicated().sum())
````

````python
print("Lead times sample:", df_clean['Lead times'].head(10).tolist())
print("Lead time sample:", df_clean['Lead time'].head(10).tolist())
print("\nAre they identical?", df_clean['Lead times'].equals(df['Lead time']))
````

````python
categorical_cols = df_clean.select_dtypes(include='object').columns
print("Categorical columns:", categorical_cols.tolist())
print("\n")

for col in categorical_cols:
    print(f"Column: {col}")
    print(f"Unique values: {df_clean[col].nunique()}")
    print(df_clean[col].value_counts())
    print("-" * 40)
````
